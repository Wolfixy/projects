```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_list = ['ruslit_full_dictionary', 'poetry_full_dictionary', 'ruslit_words', 'ruslit_lemmas', 'poetry_words', 
            'poetry_lemmas', 'ruslit_nouns', 'poetry_nouns', 'ruslit_verbs', 'poetry_verbs']

dictionaries = {}

#importing dictionaries from .cvs files
for i in csv_list:
    dictionaries[i] = pd.read_csv(i + '.csv', sep = ';', header = 'infer', nrows = 2000)

#inserting indexes
for i in csv_list:
    dictionaries[i].insert(1, 'i', range(1, len(dictionaries[i]) + 1))

#calculating theoretical frequencies and inserting them into dictionaries
for i in csv_list:
    top_frequency = dictionaries[i].iloc[0, 2]
    index = np.array(dictionaries[i].iloc[:, 1])
    for j in index:
        zipf_frequency = top_frequency * (1 / index)
    zipf_frequency = np.around(zipf_frequency)
    dictionaries[i]['Fz'] = zipf_frequency.tolist()
dictionaries['ruslit_full_dictionary']
```

```python
x1 = np.array(dictionaries['ruslit_full_dictionary'].iloc[:50, 1])
y1 = np.array(dictionaries['ruslit_full_dictionary'].iloc[:50, 2])
y1_t = np.array(dictionaries['ruslit_full_dictionary'].iloc[:50, 3])
x2 = np.array(dictionaries['poetry_full_dictionary'].iloc[:50, 1])
y2 = np.array(dictionaries['poetry_full_dictionary'].iloc[:50, 2])
y2_t = np.array(dictionaries['poetry_full_dictionary'].iloc[:50, 3])

fig1, axs = plt.subplots(1, 2, figsize = (10, 3))

axs[0].set_title("корпуса русских рассказов")
axs[0].plot(x1, y1, '--', label = "Экспериментальный результат")
axs[0].plot(x1, y1_t, label = "Теоретический результат")
axs[0].set_xlabel("Ранг")
axs[0].set_ylabel("Частота")

axs[1].set_title("корпуса поэтических текстов")
axs[1].plot(x2, y2, '--', label = "Экспериментальный результат")
axs[1].plot(x2, y2_t, label = "Теоретический результат")
axs[1].set_xlabel("Ранг")
axs[1].set_ylabel("Частота")
axs[0].legend(loc = 'upper center', bbox_to_anchor = (1, -0.2))
fig1.suptitle("Выполнение закона Ципфа для полного частотного" 
              "словаря", y = 1.05)
plt.subplots_adjust(wspace = 0.25)
plt.show()
```
